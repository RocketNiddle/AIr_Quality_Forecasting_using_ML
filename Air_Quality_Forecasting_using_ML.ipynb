**Forcasting with FB Prophet Algorithm**

*   Powerful time series forecasting algorithm that can capture complex patterns in the data such as seasonality, trends, and the effect of holidays
*   It is based on a Bayesian approach that allows for uncertainty estimation in the predictions.

ideally YYYY-MM-DD for a date or YYYY-MM-DD HH:MM:SS for a timestamp

Link : https://facebook.github.io/prophet/docs/quick_start.html#python-api



```python
#converting the date from DD/MM/YYYY to YYYY-MM-DD
#dayfirst=True extra

date_info =  pd.to_datetime(air_quality_data['Date'],dayfirst=True)
print(date_info)
```

```python
#changeing 10.00.00 to 10:00:00 using lambda function

time_info = air_quality_data['Time']

time_info = time_info.apply(lambda x:x.replace('.',':'))

print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
# Combining 2 series to a pandas dataframe

date_time = pd.concat([date_info,time_info], axis =1)

date_time.head()
```

```python
date_time.shape
```

```python
#combining date and time

date_time['ds'] = date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str)

date_time.head()
```

```python
date_time.info()

data = pd.DataFrame()


```

Covert ds from object datatype to date time format

```python
data['ds'] = pd.to_datetime(date_time['ds'])

data.head()
```

```python
data['y'] = air_quality_data['RH']
data.tail()
```

```python
from prophet import Prophet
```

```python
# training the Prophet model

model = Prophet()
model.fit(data)
```

```python
# Python
future = model.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()

```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
